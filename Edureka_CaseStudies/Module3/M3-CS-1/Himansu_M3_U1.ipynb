```python
import pandas as pd             # Pandas for managing datasets
import numpy as np
import matplotlib.pyplot as plt  # Matplotlib for additional customization
import seaborn as sns            # Seaborn for plotting and styling
%matplotlib inline
```

```python
path='C:\\Users\\hpanigr\\Documents\\Study_Docs\\Trainings\\ML_Python\\Usecase-1\\'
data = pd.read_csv(path + 'prisoners.csv')
data.columns = ['state','year','ele_edu','adult_edu','higher_edu','comp_edu']
data.columns
```

```python
data_df = pd.DataFrame(data)
#data.index[]
data_df.iloc[[1]] # row with index 1
data_df.loc[[1]] # row with index 1
```

```python
data_df[:][(data_df['ele_edu']==0) | (data_df['adult_edu']==0) | (data_df['higher_edu']==0) | (data_df['comp_edu']==0)]
```

```python
data_df['total_benefited'] = data_df[['ele_edu','adult_edu','higher_edu','comp_edu']].sum(axis=1)
data_df.tail(5)
```

```python
x, y = data_df['state'],data_df['total_benefited']
plt.figure(figsize = (15,7))
plt.xlabel("State")
plt.ylabel("Total Benefited")
plt.legend(loc="lower right")
plt.bar(x,y)
plt.xticks(rotation=-90)
plt.yticks(rotation=0)
```
